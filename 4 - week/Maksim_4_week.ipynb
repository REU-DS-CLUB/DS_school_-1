УЖЕ ВИДНО, ЧТО ПРИЗНАК RAM ЯВЛЯЕТСЯ САМЫМ ГЛАВНЫМ

```python
from sklearn.decomposition import PCA
import seaborn as sns
pca_test = PCA(n_components=20)
pca_test.fit(X_train)
sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=3, ymin=0, ymax=1)
display(plt.show())
evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr

display(pca_df.head(3))
```

получается тока 3 признака объясняют на дисперсию
причем 2 на 95% и 1 на 5% 

значит можно понизить размерность с 20 до 3

```python
pca = PCA(n_components=3)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 1000, num = 10)]
max_features = ['log2', 'sqrt']
max_depth = [int(x) for x in np.linspace(start = 1, stop = 15, num = 15)]
min_samples_split = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
min_samples_leaf = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
bootstrap = [True, False]
param_dist = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
another_model = RandomForestClassifier()
rs = RandomizedSearchCV(another_model, 
                        param_dist, 
                        n_iter = 100, 
                        cv = 3, 
                        verbose = 1, 
                        n_jobs=-1, 
                        random_state=0)
rs.fit(X_train_pca, y_train)
rs.best_params_
```

```python
rs_df = pd.DataFrame(rs.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
rs_df = rs_df.drop([
            'mean_fit_time', 
            'std_fit_time', 
            'mean_score_time',
            'std_score_time', 
            'params', 
            'split0_test_score', 
            'split1_test_score', 
            'split2_test_score', 
            'std_test_score'],
            axis=1)
rs_df.head(15)
```

на основании рандомных параметров видно, что модель хорошо себя чувствует при n_estimators около 670. Можно попробовать взять диапазон от 670 до 700 

min_samples_split = 12, 18, 23

min_samples_leaf = 2, 7

max_depth от 7 до 15

```python
parameters = { 'n_estimators' : [340, 670], 'max_features' : ['sqrt', 'log2'], 
              'max_depth' : np.arange(7, 15, 1), 'min_samples_split' : [12, 18, 23],
              'min_samples_leaf' : [2,7], 'bootstrap' : [True, False] }
```

```python
from sklearn.model_selection import GridSearchCV
new_model = RandomForestClassifier()
gs = GridSearchCV(new_model, parameters, cv = 2, verbose = 1, n_jobs=-1)
gs.fit(X_train_pca, y_train)
rfc_3 = gs.best_estimator_
gs.best_params_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = rfc_3.predict(X_test_pca)

print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
gridparams = {
    'learning_rate': [float(x) for x in np.linspace(start = 0.05, stop = 0.8, num = 10) ],
    'num_leaves': [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)],
    'boosting_type' : ['gbdt', 'dart', 'goss'],
    'objective' : ['binary'],
    'max_depth' : [2, 4, 6, 8],
    'random_state' : [113]
}

model = LGBMClassifier()
grid = RandomizedSearchCV(model, gridparams, cv=10, n_jobs = -1, n_iter=10)
grid.fit(X_train_pca,y_train)
```

```python
grid.best_params_
```

```python
rs_df = pd.DataFrame(grid.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
rs_df = rs_df.drop([
            'mean_fit_time', 
            'std_fit_time', 
            'mean_score_time',
            'std_score_time', 
            'params', 
            'split0_test_score', 
            'split1_test_score', 
            'split2_test_score', 
            'std_test_score'],
            axis=1)
rs_df.head(10)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'learning_rate': [0.133333, 0.716667],
    'num_leaves': [31, 115],
    'boosting_type' : ['goss'],
    'objective' : ['multiclass'],
    'max_depth' : [6, 8],
    'random_state' : [113]
}

model = LGBMClassifier()

lgb = GridSearchCV(model, param_grid=parameters, n_jobs=-1, cv=10)
lgb.fit(X_train_pca,y_train)
pred = lgb.predict(X_test_pca)
print(classification_report(y_test, pred))
```
