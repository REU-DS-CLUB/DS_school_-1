```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from matplotlib import pyplot as plt
from sklearn import metrics

data = pd.read_csv("Second_week_third_dataset.csv", index_col='Id')
data1 = data.drop('Species',axis=1)
pca = PCA(n_components=2).fit(data1)
pca_2d = pca.transform(data1)
dbscan = DBSCAN().fit(data1)
clast_labels = dbscan.labels_
data['clast'] = clast_labels

colour = {1: 'r', 0: 'g', -1: 'b'}
s = data['clast']==data["Species"]

col = [colour[i] for i in data['clast']]
x = [pca_2d[i, 0] for i in range(len(pca_2d))]

y = [pca_2d[i, 1] for i in range(len(pca_2d))]
plt.scatter(x,y,c=col)
```

```python
not_nan_clast = data.shape[0]-data["Species"].isna().sum()

sum_clast_spec = data['clast']==data["Species"]

sum_clast_spec.sum()/not_nan_clast
```
