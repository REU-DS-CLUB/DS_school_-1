```python
import pandas as pd 
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics

dataset = pd.read_csv("Second_week_first_dataset.csv", index_col = 'Unnamed: 0')
```

```python
dataset
```

```python
plt.scatter(dataset['X'],dataset['Y']) 
```

```python
claster = AgglomerativeClustering(linkage='single').fit(dataset)

dataset["clast"]=claster.labels_

dataset
```

```python
colour_dict = {0:'dodgerblue',1:'limegreen'}

colours = [colour_dict[i] for i in dataset.clast]
plt.scatter(dataset.X, dataset.Y, c=colours)
```
